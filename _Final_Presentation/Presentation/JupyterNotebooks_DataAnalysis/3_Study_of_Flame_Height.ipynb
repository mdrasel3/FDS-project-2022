```python
import fdsreader
import matplotlib.pyplot as plt

# define the path to the data
path_to_data_list = ['E:\CSiS_Sem2\FireSimulation\Exam2022\Raihan_Straight_Burner\Raf_Sim\Fire_Raihan\slcf_study_1',
                    'E:\CSiS_Sem2\FireSimulation\Exam2022\Raihan_Straight_Burner\Raf_Sim\Fire_Raihan\slcf_study_2']

print("Study 1: .05 X .02 X .05, Count: 312800\nStudy 2: .025 X .02 X .1, Count: 625600\n")

for path_to_data in path_to_data_list:
    sim = fdsreader.Simulation(path_to_data)

    print(sim)

    plt.plot(sim.hrr['Time'], sim.hrr['HRR'])
    plt.xlabel('time / s')
    plt.ylabel('heat release rate / kW')
    plt.grid()
    plt.show()

    # print available slice data
    for slice in sim.slices:
        print(slice.quantity, slice.extent)

    # sim.slices[sliceid][meshid].data[timestep, direction1, direction2]
    # Initial visualasation of the data
    it = sim.slices[0].get_nearest_timestep(5)
    plt.imshow(sim.slices[0][0].data[it])
    plt.show()
```
