<i>a. Observe Grid Sensitivity with equal size in three axis</i>

```python
import pandas as pd
import matplotlib.pyplot as plt

devc_csv = "data/Steckler_012_devc_18X14X11.csv"
devc_data_df = pd.read_csv(devc_csv, header=1)
temp_list = [x for x in devc_data_df.columns if 'temp-' in x]

devc_csv2 = "data/Steckler_012_devc_36X28X22.csv"
devc_data_df2 = pd.read_csv(devc_csv2, header=1)
temp_list2 = [x for x in devc_data_df2.columns if 'temp-' in x]

plt.rcParams['figure.dpi'] = 100

fig, (ax1, ax2) = plt.subplots(2)
print('Grid Sensitivity Analysis of Temperature vs Time for Steckler Compartment at Device: 1')
ax1.plot(devc_data_df["Time"], devc_data_df[temp_list[0]])
ax2.plot(devc_data_df2["Time"], devc_data_df2[temp_list2[0]])

ax1.title.set_text("18X14X11, .2m")
ax2.title.set_text("36X28X22, .1m")
fig.tight_layout()
```

<i>b. Observe Grid Sensitivity with different size in three axis</i>

```python
# We can take a certain timestamp and can assume a relative true value of T
# and for the same timestamp we can compare the T value for different combination of x, y, z (different combination of mesh)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.title.set_text("Study of different size of mesh on different axis")

x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)
c = np.random.standard_normal(100)

img = ax.scatter(x, y, z, c=c)
plt.show()
```

<i>c. Observe Grid Sensitivity with dividing the Computational domain</i>
<p>
OpenMP can be used to exploit multiple cores on a single computer. Regardless of the Computational domain being broken up into multiple meshes or not, OpenMP will work. On the other hand, MPI requires the computational domain to be divided into multiple meshes and typically each mesh is assigned its own process. We can also use OpenMP and MPI together.
</p>
<b>This is an example of how we can break the meshes differently and assign different processor to speed up our simulation</b>
<img src="img/subdomain_divided_mesh.png" width="600"/>
<p>
In this particular case, we can also have very fine meshed around the air supplier and around burner because most of the action is happening there. We may have large meshes on the other places because they are not very important in our simulation.
</p>
